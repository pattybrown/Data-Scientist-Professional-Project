## Data Validation and Cleaning

The original data contained 947 rows and 8 columns. Each column in the data was checked for duplicates, missing data, unexpected values, and appropriate data type. The provided code is annotated to provide further step-by-step detail. There were originally 52 recipes missing values for all nutrient-related (calories, carbohydrate, sugar, and protein) data. Because of the limited information offered by these data, all recipes (13 total) with missing nutrient data and low traffic were removed from the dataset prior to the analysis. The distribution of the remaining nutrient data was visualized and determined to be non-normal. Therefore, median imputation was used to handle missing values for the remaining nutrient data columns. After validation, the data contained 934 rows and 8 columns. Further validation and cleaning processes are discussed below. 

The following steps were taken for validation and cleaning of data in each column: 
- recipe: 
	- Integer as expected, no duplicate/unexpected/missing values. 
- calories: 
	- Numeric as expected. Missing data was replaced with the median. No duplicate/unexpected values. 
- carbohydrate:
	- Numeric as expected. Missing data was replaced with the median. No duplicate/unexpected values. 
- sugar:
	- Numeric as expected. Missing data was replaced with the median. No duplicate/unexpected values. 
- protein:
	- Numeric as expected. Missing data was replaced with the median. No duplicate/unexpected values. 
- category:
	- Reclassified from character to factor type. The data originally showed 11 levels, instead of the desired 10. All "Chicken Breast" values were replaced with "Chicken" to reach desired factor levels. No missing/duplicate data. 
- servings:
	- Original column was classified as character type. This column had observations with strings attached to the numeric value. The column was cleaned to keep only numeric values and reclassified as numeric type. No missing/duplicate values. 
- high_traffic:
	- Original column was character type and contained one value ("High") for indicating high-traffic recipes, and missing values (NA) for all other recipes (373). "High" values were changed to "True" and missing values were replaced with "False" for clarity. Column was reclassified as binary or factor with 2 levels. 

```r
recipe <- read.csv("recipe_site_traffic_2212.csv") # Read in the data
str(recipe)
```

```r
colSums(is.na(recipe)) # Check all columns for missing data
```

```r
recipe %>% group_by_all() %>% filter(n()>1) %>% ungroup() # Check all columns for duplicates

recipe %>% filter(is.na(calories)) # Show columns with missing nutrient data
```

```r
recipe$high_traffic <- replace_na(recipe$high_traffic, "False") # Replace all missing values with "False"
recipe$high_traffic <- str_replace(recipe$high_traffic, "High", "True") # Replace "High" values with "True"
recipe$high_traffic <- as.factor(recipe$high_traffic) # Classify as factor with 2 levels ("True", "False")
```

```r
recipe <- recipe %>% filter(!is.na(carbohydrate) | high_traffic == "True")
 # Remove all data with missing nutrient info and low traffic
```

### View nutrient data distributions before handling missing data: 

```r
carbhist <- ggplot(recipe, aes(carbohydrate)) +
  geom_histogram() 
calhist <- ggplot(recipe, aes(calories)) +
  geom_histogram()
sughist <- ggplot(recipe, aes(sugar)) +
  geom_histogram()
prothist <- ggplot(recipe, aes(protein)) +
  geom_histogram()

grid.arrange(carbhist, calhist, sughist, prothist)
```

Because all nutrient values show non-normal distributions, median imputation was selected to handle missing data in these columns. 

```r
recipe$calories <- recipe$calories %>% 
  replace_na(median(recipe$calories, na.rm = TRUE)) # Replace missing numeric values with the median

recipe$carbohydrate <- recipe$carbohydrate %>% 
  replace_na(median(recipe$carbohydrate, na.rm = TRUE)) # Replace missing numeric values with the median

recipe$sugar <- recipe$sugar %>% 
  replace_na(median(recipe$sugar, na.rm = TRUE)) # Replace missing numeric values with the median

recipe$protein <- recipe$protein %>% 
  replace_na(median(recipe$protein, na.rm = TRUE)) # Replace missing numeric values with the median

recipe$servings <- as.numeric(gsub(".*?([0-9]+).*", "\\1", recipe$servings)) # Remove text from column and classify as numeric
```

```r
recipe$category <- factor(recipe$category)
levels(recipe$category) # Check levels - has 11, should have 10
recipe$category <- str_replace(recipe$category, "Chicken Breast", "Chicken") # Combine like factors
recipe$category <- factor(recipe$category)
levels(recipe$category) # Recheck levels
colSums(is.na(recipe)) # Check all columns for missing data
```

Calculate the proportion of high-traffic recipes to low-traffic recipes by serving and by category. Plot the proportions to further visualize the differences in category and explore potential differences in serving size:

```r
high_by_serv <- recipe %>% group_by(servings) %>% filter(high_traffic == "True") %>% 
  count()
low_by_serv <- recipe %>% group_by(servings) %>% filter(high_traffic == "False") %>% 
  count()
traf_by_serv <- left_join(high_by_serv, low_by_serv, by = "servings") %>% 
  rename("True" = "n.x", "False" = "n.y")
traf_by_serv2 <- traf_by_serv %>% group_by(servings) %>% mutate(props = True / False)

ggplot(traf_by_serv2, aes(servings, props, fill = servings)) +
  geom_col() +
  labs(x = "Servings", y = "Proportions",
       title = "Proportions of High / Low Traffic Recipes by Serving Size") +
  theme(legend.title = element_blank()) +
  scale_x_continuous(breaks = c(1,2,4,6), limits = c(0.5, 6.5))
options(repr.plot.width = 15, repr.plot.height = 8)
```

Proportions by serving do not seem to indicate any significant results. This shows us that the serving size of a recipe likely does not have a large impact on website traffic. 

```r
high_by_cat <- recipe %>% group_by(category) %>% filter(high_traffic == "True") %>% 
  count() # Extract True values by category
low_by_cat <- recipe %>% group_by(category) %>% filter(high_traffic == "False") %>% 
  count() # Extract False values by category
traf_by_cat <- left_join(high_by_cat, low_by_cat, by = "category") %>% 
  rename("True" = "n.x", "False" = "n.y")
traf_by_cat2 <- traf_by_cat %>% group_by(category) %>% mutate(props = True / False) # Combine into grouped dataframe and calculate proportions of True/False for high traffic

# Plot proportions by category
ggplot(traf_by_cat2, aes(category, props, fill = category)) +
  geom_col() +
  scale_color_brewer() +
  labs(x = "Category", y = "Proportions",
       title = "Proportions of High / Low Traffic Recipes by Category") +
  theme(legend.title = element_blank(), plot.title = element_text(size = 22))
options(repr.plot.width = 15, repr.plot.height = 8)
```

Proportions by serving do indicate noticable trends. This shows us that the category seems to play an important role in consumer interest, and whether or not a recipe will result in high website traffic. Specifically, recipes featuring vegetables, potato, and pork have the highest likelihood of resulting in high traffic. Beverages offer the lowest proportion of high-traffic recipes, with breakfast and chicken recipes performing only slightly better.

## Model Development

The goal of this analysis is to classify recipes into a "True" or "False" high-traffic category. As such, this is considered a binary classification problem. Logistic regression is one of the most common and effective modeling approaches for this type of problem, and was therefore selected for the baseline model. A random forest model was selected for comparison, as these models can also be very effective for classification. Category, protein, carbohydrate, sugar, and calories were used to predict high traffic in both models. 

10-fold cross validation was used to validate each model. Rather than split the data into train and test sets, this validation method allows us to leverage all of the data and still validate the model performance using out of sample data points. 

A threshold of 0.5 was used for both models. If the predicted probability of a recipe being classified as "high traffic" is greater that 0.5, the result will be "True." If the predicted probably is less than 0.5, the result will be "False."

```r
# Set custom train control for models with 10-fold cross validation:
myControl <- trainControl(
  method = "cv",
  number = 10,
  summaryFunction = twoClassSummary, 
  classProbs = TRUE, 
  verboseIter = TRUE
)
```

Baseline model: logistic regression

```r
# Fit GLM model with caret using 10-fold cross validation
set.seed(22)
model2 <- train(
  high_traffic ~ category + protein + carbohydrate + sugar + 
    calories + servings,
  data = recipe,
  family = "binomial",
  method = "glm",
  trControl = myControl
)
summary(model2)

# Predict on full data: p2
p2 <- predict(model2, recipe, type = "prob") #predicted probabilities
recipe$prediction2 <- ifelse(p2$True > 0.5, 1, 0) #factored probabilities
```

Comparison model: Random forest

```r
# Fit random forest model with caret using 10-fold cross validation
set.seed(22)
model3 <- train(
  high_traffic ~ category + protein + carbohydrate + sugar + 
    calories + servings,
  data = recipe,
  family = "binomial",
  method = "ranger",
  trControl = myControl
)
```

```r
summary(model3)
```

```r
# Predict on full data: p3
p3 <- predict(model3, recipe, type = "prob") #predicted probabilities
recipe$prediction3 <- ifelse(p3$True > 0.5, 1, 0) #factored probabilities
```

## Model evaluation

Confusion matrices were selected to evaluate and compare the model performance. Confusion matrices are sound tools for evaluating binary classification models and are therefore fitting for this analysis. Our objective is to evaluate each model's performance in successfully predicting high-traffic recipes. Confusion matrix metrics allow us to specifically evaluate performance for true positive and true negative predictions individually, providing an opportunity to look directly at high-traffic prediction success as well as overall performance. 

```r
# Confusion Matrix for logistic regression model evaluation: 
p_class2 <- factor(ifelse(p2$True >0.5, "True", "False"))
table(p_class2, recipe$high_traffic)
confusionMatrix(p_class2, recipe$high_traffic, mode = "everything", positive = "True")
```

```r
# Confusion Matrix for random forest model evaluation: 
p_class3 <- factor(ifelse(p3$True >0.5, "True", "False"))
table(p_class3, recipe[["high_traffic"]])
confusionMatrix(p_class3, recipe$high_traffic, mode = "everything", positive = "True")
```

As seen in the results, the confusion matrices classify "True" results as the positive class and "False" results as the negative class. Because our objective is to correctly predict 80% of high-traffic ("True") occurrences, we are most concerned with the positive class results to meet this goal. Sensitivity is the measurement of the true positive rate, and is therefore the metric used to evaluate the models' ability to correctly predict high-traffic recipes. 

The F1 metric is a measure of overall model accuracy integrating both sensitivity and specificity (true negative rate). As a measure of overall accuracy, the F1 metric is used to compare performance between models. 

The logistic regression model performed well with a sensitivity of 0.80 and F1 of 0.81.

The random forest model outperformed the logistic regression model with a sensitivity of 0.93 and F1 of 0.90. These results suggest that the random forest model should be used as the preferred model for predicting high-traffic recipes. 

## Business Metrics

The objectives of this analysis were to build a predictive model that meets the following business metrics:

- Predict which recipes will lead to high traffic. 
- Correctly predict high traffic recipes 80% of the time. 

The key performance indicators (KPI) for model performance described above are sensitivity and F1. Sensitivity indicates the model's ability to successfully predict high-traffic recipes, while F1 indicates overall performance. A score of 1.0 indicates perfect performance in both metrics. 

Both the logistic regression model and the random forest model met or exceeded the KPI and business objectives with sensitivity and F1 scores equal to 80% or higher. The random forest model performed especially well both metrics meeting or exceeding 90%. 

These results suggest that the random forest model successfully predicts high-traffic recipes 93% of the time, and has an overall balanced accuracy of 90%. The random forest model should be utilized to classify new recipes using the following as an example: 

```r
set.seed(23)
new_recipe <- slice_sample(recipe, n = 1) # Select random recipe from dataset to act as new recipe

predict(model3, newdata = new_recipe, type = "prob") # Run model on new recipe
ifelse(predict(model3, newdata = new_recipe, type = "prob") > 0.5, 1, 0) # Factor result
```

These results suggest that the "new" recipe has a 76% probability of resulting in high traffic. Because this probability is greater than our 50% threshold, the result is "True" and the recipe is classified as "High traffic."
